```python
from Libraries import LinearRegression
import pandas as pd
```

```python
csv = "Data/CSV/"
file = "Hour_Score"
# file = "Exp_Salary"
data_path = csv + file + ".csv"
df = pd.read_csv(data_path)

linear = LinearRegression.LinearRegressionPipeline(df)
```

```python
# 1.0 Khám phá dữ liệu
linear.explore()
```

```python
# 2.0 Hồi quy tuyến tính (đơn biến) - Huấn Luyện
linear.fit_univariate()
```

```python
# 2.1 Hồi quy tuyến tính (đơn biến) - Diễn Giải
linear.interpret_univariate()
```

```python
# 3.0 Hồi quy tuyến tính (đa biến)
linear.fit_multivariate_ols(show_summary=True)
```

```python
# 4.0 Đánh giá mô hình đa biến (tự tính MAE, RMSE, R²)
linear.evaluate_multivariate()
```

```python
# Dự đoán giá trị mới
predict = linear.predict_multivariate([9.567])
print (predict[0])
```
